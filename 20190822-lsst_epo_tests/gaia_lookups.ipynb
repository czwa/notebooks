```python
from astroquery import gaia
import numpy as np
```

```python
#https://gea.esac.esa.int/archive/
#https://gea.esac.esa.int/archive/documentation/GDR2/Catalogue_consolidation/chap_cu9val_cu9val/ssec_cu9xma/sssec_cu9xma_extcat.html
S = gaia.GaiaClass()
```

```python
# M2
R = S.cone_search("323.36258, -0.82325", "15 arcminute")
```

```python
colnames = ['dist', 'source_id', 'ra', 'dec', 'parallax', 'pmra', 'pmdec', 
          'phot_g_mean_mag', 'phot_rp_mean_mag', 'phot_bp_mean_mag',
          ]
```

```python
indices_I_want = np.isfinite(R.results['parallax'])
indices_I_want = np.logical_and(indices_I_want, np.isfinite(R.results['phot_g_mean_mag']))
x = R.results[indices_I_want][colnames]
```

```python
x
```

```python
%matplotlib notebook
import matplotlib.pyplot as plt
plt.scatter(x['pmra'], x['pmdec'], s=0.5)
plt.xlim(-10, 10)
plt.ylim(-10, 10)
```
